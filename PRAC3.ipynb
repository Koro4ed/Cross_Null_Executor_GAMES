```python
import tkinter as tk
from tkinter import messagebox

# Создаем окно приложения
window = tk.Tk()
window.title("Крестики-нолики")

# Определяем переменные
current_player = "X"
game_over = False
board = [["" for _ in range(3)] for _ in range(3)]

# Функция для обработки клика по ячейке
def cell_click(row, col):
    global current_player, game_over

    # Проверяем, что ячейка пустая и игра не окончена
    if board[row][col] == "" and not game_over:
        board[row][col] = current_player
        button = buttons[row][col]
        button.config(text=current_player)

        # Проверяем условия окончания игры
        if check_winner(current_player):
            messagebox.showinfo("Игра окончена", f"Игрок {current_player} победил!")
            game_over = True
        elif check_draw():
            messagebox.showinfo("Игра окончена", "Ничья!")
            game_over = True
        else:
            # Меняем текущего игрока
            current_player = "O" if current_player == "X" else "X"

# Функция для проверки победителя
def check_winner(player):
    # Проверяем строки
    for row in range(3):
        if board[row][0] == board[row][1] == board[row][2] == player:
            return True

    # Проверяем столбцы
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col] == player:
            return True

    # Проверяем диагонали
    if board[0][0] == board[1][1] == board[2][2] == player:
        return True
    if board[0][2] == board[1][1] == board[2][0] == player:
        return True

    return False

# Функция для проверки ничьей
def check_draw():
    for row in range(3):
        for col in range(3):
            if board[row][col] == "":
                return False
    return True

# Создаем кнопки для ячеек
buttons = []
for row in range(3):
    button_row = []
    for col in range(3):
        button = tk.Button(window, text="", width=10, height=5,
                           command=lambda r=row, c=col: cell_click(r, c))
        button.grid(row=row, column=col, padx=5, pady=5)
        button_row.append(button)
    buttons.append(button_row)

# Запускаем главный цикл приложения
window.mainloop()
```

```python
# Создаем окно приложения
window = tk.Tk()
window.title("Игра в палачи")

# Определяем переменные
words = ["apple", "banana", "cherry", "date", "elderberry"]  # Предопределенные слова
current_word = ""
display_word = ""
wrong_guesses = 0

# Функция для начала новой игры
def new_game():
    global current_word, display_word, wrong_guesses

    # Выбираем случайное слово из списка
    current_word = random.choice(words)
    display_word = "_" * len(current_word)
    wrong_guesses = 0

    # Обновляем интерфейс
    word_label.config(text=display_word)
    wrong_label.config(text="Неправильные попытки: 0")
    input_entry.delete(0, tk.END)

# Функция для обработки догадки пользователя
def make_guess():
    global display_word, wrong_guesses

    guess = input_entry.get().lower()
    input_entry.delete(0, tk.END)

    if guess in current_word:
        # Обновляем отображение угаданных букв
        new_display = ""
        for i in range(len(current_word)):
            if current_word[i] == guess:
                new_display += guess
            else:
                new_display += display_word[i]
        display_word = new_display
        word_label.config(text=display_word)

        # Проверяем, если слово угадано полностью
        if display_word == current_word:
            messagebox.showinfo("Поздравляем!", "Вы угадали слово!")
            new_game()
    else:
        wrong_guesses += 1
        wrong_label.config(text="Неправильные попытки: " + str(wrong_guesses))

        # Проверяем, если игра проиграна
        if wrong_guesses >= 6:
            messagebox.showinfo("Палач", "Вы проиграли! Загаданное слово: " + current_word)
            new_game()

# Создаем элементы интерфейса
word_label = tk.Label(window, text=display_word, font=("Arial", 24))
word_label.pack(pady=10)

input_frame = tk.Frame(window)
input_frame.pack(pady=10)

input_label = tk.Label(input_frame, text="Введите букву:")
input_label.grid(row=0, column=0)

input_entry = tk.Entry(input_frame)
input_entry.grid(row=0, column=1)

guess_button = tk.Button(window, text="Угадать", command=make_guess)
guess_button.pack(pady=10)

wrong_label = tk.Label(window, text="Неправильные попытки: 0")
wrong_label.pack(pady=10)

new_game()  # Начинаем новую игру при запуске программы

# Запускаем главный цикл приложения
window.mainloop()
```
